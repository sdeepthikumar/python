```python
import requests
from bs4 import BeautifulSoup
import datetime
```

```python

homepage = "https://www.imdb.com/movies-coming-soon/"

date = datetime.datetime(2020,3,1)

for x in range(0,12):
    next_date = date + datetime.timedelta(days=31)
    next_month = next_date.strftime('%Y-%m')
    date = datetime.datetime(next_date.year,next_date.month,next_date.day)
    link = homepage + next_month
    
    response = requests.get(link)
    soup = BeautifulSoup(response.content,"html.parser")
    
    cards = soup.find_all("td",attrs={"class":"overview-top"})
    
    for card in cards:
        name = card.find("a")
        print(name.text)
```
