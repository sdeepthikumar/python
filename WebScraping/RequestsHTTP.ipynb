```python
count =0
response = requests.get('https://api.covid19api.com/summary')
covid_data_json = response.json()
for eachdict in covid_data_json['Countries']:
    if eachdict['NewConfirmed']>10000:
        count += 1
        print(eachdict['Country'],eachdict['NewConfirmed'])
print("Count = ",count)
```

```python
count =0
response = requests.get('https://api.covid19api.com/summary')
covid_data_json = response.json()
for eachdict in covid_data_json['Countries']:
    percentage_recovered = (eachdict['TotalRecovered']/eachdict['TotalConfirmed'])*100
    if percentage_recovered ==100:
        count += 1
        print(eachdict['Country'],percentage_recovered)
print("Count = ",count)
```
