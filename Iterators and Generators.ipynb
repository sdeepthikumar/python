```python

def fetch_data(fp):
    output = []
    count = 0
    while (True):
        count +=1
        lines = fp.readline()
        output.append(lines.strip())
        
        yield output
        return output
        output = []
        count = 0

fp = open("sample_text.txt", "r", encoding = "utf-8")
#print(fetch_data(fp))
g = fetch_data(fp)
itertools.tee(g,5)

print(next(g))
print(next(g))

```
